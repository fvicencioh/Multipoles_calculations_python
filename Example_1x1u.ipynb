```python
import numpy as np

bempp_solution = np.array([-5499.546225, -5319.362699, -5217.354216])
pygbe_solution = np.array([-5562.036124, -5368.941416, -5262.233235])
```

```python
import matplotlib.pyplot as plt

rho = np.array([1,  1.125, 1.25])
plt.plot(rho, bempp_solution, marker="x", label="Bempp")
plt.plot(rho, pygbe_solution, marker="o", label="Pygbe")

plt.xlabel("Grid Scale")
plt.ylabel("Solution")
plt.title("Comparison")
plt.legend()
```

```python
error = np.zeros(3)
for i in range(3):
    error[i] = np.abs(np.sqrt((bempp_solution[i]-pygbe_solution[i])**2)/pygbe_solution[i])
plt.loglog((rho**2)/2, error, marker="x", label="Difference")
plt.xlabel("Area")
plt.ylabel("Difference")
plt.title("Comparison")
plt.legend()
```
